```python
import pandas as pd
```

### T-sherpa
- 데이터 불러오기

```python
df = pd.read_csv('./ReviewClassification/data/tsherpa_all.csv')
```

- 제목과 내용을 하나의 컬럼으로 합치기

```python
comment_list = df['contents']
```

```python
df2 = df.copy()
```

```python
df2['title_comments'] = df2['title'] + ' ' +  df2['contents']
```

```python
df2.to_csv('sherpa_tc.csv', index = False)
```

```python
df3 = df2[df2['title_comments'].apply(lambda x: False if len(x) > 512 else True)]
```

```python
df3 = df3.reset_index(drop=True).drop([2595,2596,2597,2598,2599], axis = 0)
```

```python
comment_list_t = df3['title_comments']
```

- 합친 컬럼을 바탕으로 감성분석     
    - 사전학습된 모델 이용

```python
from transformers import pipeline
classifier = pipeline("text-classification", model="beomi/KcELECTRA-base-v2022")

emot_list_t = []

for i in range(0, len(comment_list_t)):
    preds = classifier(comment_list_t[i], return_all_scores=True)
    is_positive = preds[0][1]['score']
    emot_list_t.append(is_positive) > 0.5
```

```python
emot_list_t
```

```python
df4 = df3.copy()

df4['emotion_analysis'] = emot_list_t
```

```python
df4['emotion_analysis'] = df4['emotion_analysis'].apply(lambda x: True if x > 0.5 else False)
```

```python
df4.to_csv('sherpa_emot.csv', index = False)
```

# 아이스크림

- 불러오기

```python
DF = pd.read_csv('./data/아이스크림 연수후기.csv').dropna()
DF2 = DF.rename(columns = {'0': 'class', '1': 'title', '2': 'rated', '3': 'comments', '4': 'classified', '5': 'date'})
```

```python
DF2 = DF2.reset_index(drop=True)
```

- 제목과 내용 합치기

```python
DF2['title_comments'] = DF2['title'] + ' ' +  DF2['comments']
```

```python
DF2.to_csv('icecream.csv', index = False)
```

```python
DF3 = DF2[DF2['title_comments'].apply(lambda x: False if len(x) > 512 else True)]
```

```python
DF3 = DF3.reset_index(drop=True)
```

```python
comment_list = DF3['title_comments']
```

- 감성 분석

```python
from transformers import pipeline
classifier = pipeline("text-classification", model="matthewburke/korean_sentiment")

emot_list = []

for i in range(0, len(comment_list)):
    preds = classifier(comment_list[i], return_all_scores=True)
    is_positive = preds[0][1]['score'] > 0.5
    emot_list.append(is_positive)
```

```python
DF4 = DF3.copy()

DF4['emotion_analysis'] = emot_list
```

```python
DF4.to_csv('icecream_all_emotion.csv', index=False)
```
